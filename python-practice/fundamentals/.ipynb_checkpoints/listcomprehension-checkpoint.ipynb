```python
'''
List comprehension is the compact way of creating a list from iterable
Take each item -> Transform it -> collect to a new list
'''
```

```python
#Example -> length of each word
words = ["apple", "banana", "orange"]
lens = [word[0] for word in words]
print(lens)
```

```python
#With condition(Filtering)
evens =[n for n in range(0,21) if n%2 == 0]
print(evens)
```

```python
# extracting vowel from a string
text = "andrina"
vowel_char = [ch for ch in text if ch in 'aeiou']
print(vowel_char)
"-".join(vowel_char) #converts list to string
```

```python
#Dictionary comprehension
```

```python
data = [
    ["apple", "banana"],
    ["kiwi", "banana", "apple"],
    ["mango"]
]

flat = [word for row in data for word in row]
counted_dict = {word:flat.count(word) for word in set(flat)}

#loop way
counted_dict = {}

for row in data:
    for word in row:
        total = 0
        for r in data:
            total += r.count(word)
        counted_dict[word] = total
```

```python
print(counted_dict)
```

```python
# Set Comprehension: Unique Normalized Tokens
sentences = [
    "Hello world",
    "hello Python world",
    "WORLD of PYTHON"
]

'''Task:  Using one set comprehension, produce a set of unique lowercase words across all sentences.
Output example shape:{"hello", "world", "python", "of"}'''
```

```python
clower =[ch.lower() for ch in sentences]
print(clower)
unique = [ch.split( ) for ch in clower]
print(unique)
falt = [word for row in unique for word in row]
print(falt)
final = {word for word in falt}
print(final)

#one liner code
{word for sentence in sentences for word in sentence.lower().split()}
```

```python
#Nested Comprehensions (Flattening + Matrix Ops)
#Flatten the matrix
matrix = [[1, 2], [3, 4], [5, 6]]
flat_matrix = [num for row in matrix for num in row]
print(flat_matrix)
```

```python
# Double each element
matrix = [[1, 2], [3, 4], [5, 6]]
doubled = [[num*2 for num in row] for row in matrix]
print(doubled)
```

```python
#Create coordinate pairs (row_index, col_index)
'''
[
  [1, 2],     # row 0
  [3, 4],     # row 1
  [5, 6]      # row 2
]
'''
coords = []
for i, row in enumerate(matrix):
    for j, num in enumerate(row):
        coords.append((i, j))
#list comprehension way
[(i,j) for i,row in enumerate(matrix) for j,num in enumerate(row)]
```

```python
#Real‑World Nested Comprehension Practice
data = [[" apple ", " banana "], [" kiwi ", " mango "]]
cleaned_data = [[ch.strip() for ch in row ]for row in data]
print(cleaned_data)
```
