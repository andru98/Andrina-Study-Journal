## 1. What is List Comprehension?

A compact way to create a list from an iterable.

```
[expression for item in iterable if condition]
```

**Mental model:**
```
Take each item → (optionally filter it) → Transform it → Collect into a new list
```

| | Loop way | Comprehension way |
|---|---|---|
| Readability | Verbose | Concise |
| Speed | Slower | Faster (optimized in CPython) |
| Use case | Complex logic | Simple transforms & filters |

> **Rule:** If your loop body is more than 2 lines or has nested logic, use a regular loop. Comprehensions are for clarity, not cleverness.

## 2. Basic Syntax


````
# ── Basic: first letter of each word ──
words = ["apple", "banana", "orange"]
first_letters = [word[0] for word in words]
print("First letters:", first_letters)   # ['a', 'b', 'o']

# ── Transform: uppercase each word ──
uppercased = [word.upper() for word in words]
print("Uppercased:", uppercased)

# ── Transform: length of each word ──
lengths = [len(word) for word in words]
print("Lengths:", lengths)
````

### Conditional Transform — `if/else` inside comprehension

When you want to **transform** based on a condition (not filter), 
the `if/else` goes **before** the `for`:

```python
[value_if_true if condition else value_if_false for item in iterable]
```


````
nums = [1, 2, 3, 4, 5, 6]

# ── Label each number as even or odd ──
labels = ["even" if n % 2 == 0 else "odd" for n in nums]
print("Labels:", labels)   # ['odd', 'even', 'odd', 'even', 'odd', 'even']

# ── Nested ternary: multiple conditions ──
grades_raw = [92, 78, 88, 55, 95]
grades = ["A" if s >= 90 else "B" if s >= 80 else "C" if s >= 70 else "F"
          for s in grades_raw]
print("Grades:", grades)
````

## 3. Filtering with Conditions

When you want to **keep only some items**, the `if` goes **after** the `for`:

```python
[expression for item in iterable if condition]
```


````
# ── Filter: even numbers only ──
evens = [n for n in range(0, 21) if n % 2 == 0]
print("Evens:", evens)

# ── Filter: extract vowels from a string ──
text = "andrina"
vowels = [ch for ch in text if ch in "aeiou"]
print("Vowels:", vowels)
print("Joined:", "-".join(vowels))   # converts list back to string

# ── Filter + Transform together ──
words = ["apple", "banana", "kiwi", "mango", "fig"]
long_upper = [w.upper() for w in words if len(w) > 4]
print("Long words uppercased:", long_upper)
````

## 4. Nested Comprehensions

### 4.1 Flattening a 2D list


````
# ── Flatten matrix → 1D ──
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]
print("Flat:", flat)   # [1, 2, 3, 4, 5, 6]

# Read as: "for each row in matrix, for each num in that row, take num"
````

### 4.2 Transforming each element in a 2D list


````
# ── Double each element (preserves 2D structure) ──
matrix = [[1, 2], [3, 4], [5, 6]]
doubled = [[num * 2 for num in row] for row in matrix]
print("Doubled:", doubled)   # [[2,4], [6,8], [10,12]]

# Inner comprehension runs first → processes each row
# Outer comprehension collects the processed rows
````

### 4.3 Coordinate pairs with `enumerate()`


````
matrix = [[1, 2], [3, 4], [5, 6]]

# ── Loop way ──
coords_loop = []
for i, row in enumerate(matrix):
    for j, num in enumerate(row):
        coords_loop.append((i, j))

# ── Comprehension way ──
coords = [(i, j) for i, row in enumerate(matrix) for j, num in enumerate(row)]
print("Coordinates:", coords)
# [(0,0), (0,1), (1,0), (1,1), (2,0), (2,1)]
````

### 4.4 Real-world: cleaning whitespace from nested data


````
# ── Strip whitespace from each item in a 2D list ──
# Common when reading messy CSVs or scraped data
data = [[" apple ", " banana "], [" kiwi ", " mango "]]
cleaned = [[item.strip() for item in row] for row in data]
print("Cleaned:", cleaned)
````

## 5. Dictionary Comprehension

```python
{key: value for item in iterable}
{key: value for key, value in dictionary.items()}
```


````
# ── Word frequency counter ──
data = [
    ["apple", "banana"],
    ["kiwi", "banana", "apple"],
    ["mango"]
]

# Step 1: flatten
flat = [word for row in data for word in row]

# Step 2: count each unique word
counted = {word: flat.count(word) for word in set(flat)}
print("Word counts:", counted)
# {'apple': 2, 'banana': 2, 'kiwi': 1, 'mango': 1}
````

````
# ── Iterating over a dict with .items() ──
students = {"Anna": [85, 90, 78], "Bob": [70, 65, 80]}

# Average score per student
averages = {name: round(sum(scores) / len(scores), 2)
            for name, scores in students.items()}
print("Averages:", averages)
# {'Anna': 84.33, 'Bob': 71.67}
````

````
# ── Filtering a dict ──
scores = {"Anna": 92, "Bob": 55, "Cara": 88, "Dan": 40}
passed = {name: score for name, score in scores.items() if score >= 60}
print("Passed:", passed)
````

## 6. Set Comprehension

Same as list comprehension but with `{}` — automatically removes duplicates.

```python
{expression for item in iterable}
```


````
sentences = [
    "Hello world",
    "hello Python world",
    "WORLD of PYTHON"
]

# ── Step-by-step breakdown ──
lowered  = [s.lower() for s in sentences]
split    = [s.split() for s in lowered]
flat     = [word for row in split for word in row]
unique   = {word for word in flat}
print("Step by step:", unique)

# ── One-liner ──
unique_words = {word for sentence in sentences for word in sentence.lower().split()}
print("One-liner:", unique_words)
# {'hello', 'world', 'python', 'of'}
````

## 7. Generator Expressions

Like list comprehension but **memory-efficient** — computes one value at a time.

| | List Comprehension | Generator Expression |
|---|---|---|
| Syntax | `[x for x in list]` | `(x for x in list)` |
| Memory | Stores all values | One value at a time |
| Speed | Fast (all at once) | Slightly slower per item |
| Use when | Need to reuse the list | Just need to iterate once |

> **Key rule:** If you're passing directly into `sum()`, `max()`, `min()`, `any()`, `all()` — use a generator. No need to build the full list first.


````
import sys

big_list = range(1_000_000)

# ── List comprehension — stores all 500k values in memory ──
squares_list = [x**2 for x in big_list if x % 2 == 0]
print(f"List size in memory: {sys.getsizeof(squares_list):,} bytes")

# ── Generator — computes one at a time, no storage ──
squares_gen = (x**2 for x in big_list if x % 2 == 0)
print(f"Generator size in memory: {sys.getsizeof(squares_gen)} bytes")

# ── Using generator directly in sum() ──
total = sum(x**2 for x in big_list if x % 2 == 0)
print(f"Sum of squares of evens: {total:,}")
````

## 8. Interview Exercises

These 5 exercises cover the most common comprehension patterns tested in interviews.


### Exercise 1 — Conditional Transform


````
# Label each number as "even" or "odd" — one comprehension
nums = [1, 2, 3, 4, 5, 6]
labels = ["even" if n % 2 == 0 else "odd" for n in nums]
print(labels)
# ['odd', 'even', 'odd', 'even', 'odd', 'even']

# Key: if/else goes BEFORE the for when transforming (not filtering)
````

### Exercise 2 — `zip()` Comprehension


````
# Pair each name with their letter grade — one comprehension
names  = ["Anna", "Bob", "Cara"]
scores = [92, 78, 88]

result = [(name, "A" if score >= 90 else "B" if score >= 80 else "C")
          for name, score in zip(names, scores)]
print(result)
# [('Anna', 'A'), ('Bob', 'C'), ('Cara', 'B')]

# Key: zip() pairs two lists element by element
````

### Exercise 3 — Generator Expression


````
# Sum of squares of all even numbers up to 1,000,000
# Use a generator — never build the full list
big_list = range(1_000_000)
total = sum(x**2 for x in big_list if x % 2 == 0)
print(f"Result: {total:,}")

# Key: sum(generator) — no brackets needed inside sum()
````

### Exercise 4 — Nested Dict Comprehension


````
# Calculate average score per student
students = {"Anna": [85, 90, 78], "Bob": [70, 65, 80]}
averages = {name: round(sum(scores) / len(scores), 2)
            for name, scores in students.items()}
print(averages)
# {'Anna': 84.33, 'Bob': 71.67}

# Key: .items() gives (key, value) pairs from a dict
````

### Exercise 5 — `enumerate()` Comprehension


````
# Map index → uppercase word
words = ["python", "is", "powerful"]
indexed = {i: word.upper() for i, word in enumerate(words)}
print(indexed)
# {0: 'PYTHON', 1: 'IS', 2: 'POWERFUL'}

# Key: enumerate() gives (index, value) pairs
````

## 9. Quick Reference Cheatsheet

```python
# ── LIST COMPREHENSION ──────────────────────────────────────
[expr for x in iterable]                    # basic
[expr for x in iterable if condition]       # filter
[x if cond else y for x in iterable]        # conditional transform
[expr for row in matrix for x in row]       # flatten nested

# ── DICT COMPREHENSION ──────────────────────────────────────
{k: v for x in iterable}                    # basic
{k: v for k, v in dict.items()}             # from existing dict
{k: v for k, v in dict.items() if cond}     # filtered dict

# ── SET COMPREHENSION ───────────────────────────────────────
{expr for x in iterable}                    # unique values only

# ── GENERATOR EXPRESSION ────────────────────────────────────
(expr for x in iterable)                    # lazy evaluation
sum(expr for x in iterable if cond)         # use directly in sum/max/min

# ── USEFUL BUILT-INS TO COMBINE ─────────────────────────────
zip(a, b)           # pair two lists element by element
enumerate(lst)      # (index, value) pairs
dict.items()        # (key, value) pairs from dict
set(lst)            # remove duplicates
"".join(lst)        # list of strings → single string
str.split()         # string → list of words
str.strip()         # remove whitespace
str.lower/upper()   # case conversion
```

### ⚠️ Common Mistakes

| Mistake | Wrong | Correct |
|---|---|---|
| `if/else` position | `[x for x in lst if x>0 else 0]` | `[x if x>0 else 0 for x in lst]` |
| Nested order | `[x for x in row for row in matrix]` | `[x for row in matrix for x in row]` |
| Generator in sum | `sum([x**2 for x in lst])` | `sum(x**2 for x in lst)` |
| Dict iteration | `for x in dict` | `for k, v in dict.items()` |
